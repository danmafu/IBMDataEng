## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1. Changes the `Market Cap (US$ Billion)` column from USD to GBP
2. Rounds the Market Cap (US$ Billion)` column to 3 decimal places
3. Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python
import pandas as pd

def transform(extracted_data, exchange_rate_file):
    # Load the exchange rates DataFrame
    exchange_rates = pd.read_csv(exchange_rate_file)
    
    # Find the exchange rate of USD to GBP
    exchange_rate = exchange_rates.loc[exchange_rates['Currency'] == 'GBP', 'Rates'].values[0]
    
    # Change Market Cap from USD to GBP
    extracted_data['Market Cap (US$ Billion)'] = extracted_data['Market Cap (US$ Billion)'] * exchange_rate
    
    # Round Market Cap to 3 decimal places
    extracted_data['Market Cap (US$ Billion)'] = extracted_data['Market Cap (US$ Billion)'].round(3)
    
    # Rename column to Market Cap (GBP$ Billion)
    extracted_data.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'}, inplace=True)
    
    return extracted_data

# Example usage:
extracted_data = pd.DataFrame({'Name': ['Bank A', 'Bank B'],
                               'Market Cap (US$ Billion)': [100, 200]})

exchange_rate_file = extracted_data_gbp

# Call the transform() function
transformed_data = transform(extracted_data, exchange_rate_file)

# Print the transformed data
print(transformed_data)
```
